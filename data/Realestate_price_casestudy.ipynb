```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load dataset
file_path = r"C:\Users\Aksh\OneDrive\Attachments\real_estate_case_study December.csv"
df = pd.read_csv(file_path)
```

```python
#  2. Distribution of Prices
plt.figure(figsize=(10,6))
sns.histplot(df['Price_SEK'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Property Prices')
plt.xlabel('Price (SEK)')
plt.ylabel('Count')
plt.show()
```

```python
#  3. Average Price by City
plt.figure(figsize=(10,6))
df.groupby('City')['Price_SEK'].mean().sort_values().plot(kind='bar', color='orange')
plt.title('Average Property Price by City')
plt.xlabel('City')
plt.ylabel('Average Price (SEK)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  4. Days on Market Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Days_on_Market'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Days on Market')
plt.xlabel('Days on Market')
plt.ylabel('Number of Properties')
plt.show()
```

```python
#  6. Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
